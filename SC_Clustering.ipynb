## <b>D.2. <span style='color:#3A6D8C'>Missing Value</span></b>

Missing value atau nilai yang hilang adalah data yang tidak tersedia dalam dataset.

Penanganan Missing Values:
* `Menghapus`: Menghapus baris atau kolom dengan nilai hilang.
* `Imputasi`: Mengisi nilai hilang dengan nilai lain, seperti rata-rata, median, atau modus.
* `Modeling`: Menggunakan algoritma yang dapat menangani nilai hilang secara langsung.

### <b><span style='color:#55679C'>Quest D.2.1</span></b>

<i>Dari dataframe `data_sust_asia`, periksa dan hitung banyaknya missing value pada data, jika ditemukan maka lakukan cara mengatasinya!</i>

```python
# Buat dataframe kosong
data_kosong = pd.DataFrame(index = data_sust_asia.columns.to_list())

# Lakukan perhitungan
for negara in data_sust_asia['Country Name'].unique():
    condition = (data_sust_asia['Country Name'] == negara)
    list_banyak_null = list()
    for col in data_sust_asia.columns.to_list():
        banyak_nul = data_sust_asia.loc[condition][col].isna().sum()
        list_banyak_null.append(banyak_nul)
    data_kosong[negara] = list_banyak_null

# Buat highlight merah untuk missing value > 0
styled_df = data_kosong.T.style.map(lambda val : 'background-color: red' if val > 0 else '')

display(styled_df)
```

# <b>E. <span style='color:#3A6D8C'>Feature Scaling</span></b>

Feature scaling adalah teknik dalam praproses data yang digunakan untuk mengubah skala fitur (variabel) sehingga memiliki rentang yang seragam. Scaler yang umum digunakan adalah Min-Max Scaler, Standar Scaler, Robust Scaler, dsb.<br><br>


Salah satu cara untuk scaling data adalah StandarScaler yang dirumuskan

\begin{equation}
X' = \frac{X - \mu}{\sigma}
\end{equation}

docs : https://scikit-learn.org/1.5/modules/generated/sklearn.preprocessing.StandardScaler.html

### <b><span style='color:#55679C'>Quest E.1.1</span></b>

Lakukan proses Standard Scaling pada kolom yang telah ditentukan dengan scikit learn!

```python
# Import library yang dibutuhkan
from sklearn.preprocessing import StandardScaler

# Definisikan kolom yang akan di scaling
X = data_sust_asia[[
    'GDP (current US$) - NY.GDP.MKTP.CD',
    'Final consumption expenditure (% of GDP) - NE.CON.TOTL.ZS',
    'Gross savings (% of GDP) - NY.GNS.ICTR.ZS'
]]

# Buat objek MinMaxScaler
scaler = StandardScaler()

# Terapkan Min-Max Scaling
scaled_data = scaler.fit_transform(X)

# Buat DataFrame baru dari data yang telah diskalakan
scaled_df = pd.DataFrame(scaled_data, columns = X.columns + ' normalize')

# Tampilkan datanya
display(scaled_df.head())
```

# <b>F. <span style='color:#3A6D8C'>Modelling</span></b>

### <b><span style='color:#55679C'>Quest F.1.1</span></b>
Dengan silhouette_score tentukan jumlah cluster terbaik

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Definisikan kolom yang akan di clustering
X = data_sust_asia[[
    'GDP (current US$) - NY.GDP.MKTP.CD normalize',
    'Final consumption expenditure (% of GDP) - NE.CON.TOTL.ZS normalize',
    'Gross savings (% of GDP) - NY.GNS.ICTR.ZS normalize'
]]

k_range = range(2, 16)
silhouette = list()

# Melakukan K-Means dan menghitung Silhouette Score
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    cluster_labels = kmeans.fit_predict(X)
    score = silhouette_score(X, cluster_labels)
    silhouette.append(score)
```

```python
import plotly.express as px

# Plot grafik
fig = px.line(
    x = k_range,
    y = silhouette,
    color_discrete_sequence = ['#FFAD60']
)

fig.update_layout(
    width = 1200,
    height = 500,
    showlegend = False,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    title = dict(
        text = "<b>Nilai <span style='color:#FFAD60'>Silhouette Score</span>\
        </b><br><sup><sup>dengan Algoritma K-Means</sup></sup>",
        font = dict(
            size = 25,
            color = '#757882'
        ),
    ),
    xaxis = dict(
        showgrid = False,
        showline = True,
        linecolor = 'black',
        showticklabels = True,
        zeroline = True,
    ),
    yaxis = dict(
        showgrid = False,
        showline = True,
        linecolor='black',
        showticklabels = True,
        zeroline = True,
    ),
    margin = dict(
        l = 50
    )
)

fig.show(renderer = 'colab')
```

### <b><span style='color:#55679C'>Quest F.1.2</span></b>
Lakukan modelling dengan K-Means untuk jumlah cluster terbaik

```python
kmeans = KMeans(n_clusters = 4, random_state = 0)
cluster_labels = kmeans.fit_predict(X)

data_sust_asia['cluster'] = cluster_labels
```

```python
import plotly.graph_objects as go

# Membuat plot 3D
fig = go.Figure(
    data = [
        go.Scatter3d(
            x = data_sust_asia['GDP (current US$) - NY.GDP.MKTP.CD'],
            y = data_sust_asia['Final consumption expenditure (% of GDP) - NE.CON.TOTL.ZS'],
            z = data_sust_asia['Gross savings (% of GDP) - NY.GNS.ICTR.ZS'],
            mode = 'markers',
            marker = dict(
                size = 5,
                color = data_sust_asia['cluster'],
                opacity = 0.5,
                colorscale='Spectral',
                colorbar = dict(title='Color Legend')
            ),
            hovertext = data_sust_asia['Country Name'],
            hovertemplate='<b>%{hovertext}</b><br>GDP: %{x}<br>Final Consumption Expenditure: %{y}<br>Gross Savings: %{z}'
        )
    ]
)


fig.update_layout(
    width = 800,
    height = 800,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    title = dict(
        text = '<b>Country Segmentation by GDP</b>',
        font = dict(
            size = 20
        )
    ),
    scene = dict(
        xaxis_title = 'GDP',
        yaxis_title = 'Final Consumption Expenditure',
        zaxis_title = 'Gross Savings',
        xaxis=dict(showgrid=False),
        yaxis=dict(showgrid=False),
        zaxis=dict(showgrid=False)
    )
)

# Tampilkan plot
fig.show()
```
